```python
# ðŸ“ 1. CrewAI Example
from crewai import Agent, Task, Crew, LLM

llm = LLM(
    model="ollama/llama3",
    base_url="http://localhost:11434",
    temperature=0
)

# Defining an agent
researcher = Agent(
    role="Researcher",
    goal="Analyze LLMs for pros and cons",
    backstory="You are an expert at comparing AI models.",
    llm=llm,
    verbose=True,
    allow_delegation=False
)

# Defining a task for the agent
task = Task(
    description="Compare GPT-4, Claude, and Mistral in terms of accuracy and openness",
    expected_output="Bullet list of pros and cons per model",
    agent=researcher
)

# Creating a crew with the agent and task
crew = Crew(
    agents=[researcher],
    tasks=[task],
    verbose=True
)
crew.kickoff()
```

```python
# ðŸ“ 2. LangChain + ReAct Example
from langchain.agents import create_react_agent, AgentExecutor
from langchain_community.chat_models import ChatOllama
from langchain.tools import tool
from langchain_core.prompts import PromptTemplate


@tool
def web_search(query: str) -> str:
    """Search the web for information."""
    return f"[MOCK SEARCH RESULT] Info about: {query}"


llm = ChatOllama(
    model="llama3",
    base_url="http://localhost:11434",
    temperature=0
)


prompt = PromptTemplate.from_template(
    """You are a helpful AI agent.

You may use tools if needed.

You have access to the following tools:
{tools}

The names of the tools you can use are:
{tool_names}

Use the following format:

Question: {input}
Thought: {agent_scratchpad}
Final Answer:
"""
)


agent = create_react_agent(llm, [web_search], prompt)

agent_executor = AgentExecutor(
    agent=agent,
    tools=[web_search],
    verbose=True,
    handle_parsing_errors=True
)

agent_executor.invoke({"input": "Can you compare GPT-4, Claude, and Mistral?"})

```

```python
# ðŸ“ 3. AutoGen Example - Turn-based Chat
from autogen import AssistantAgent, UserProxyAgent

llm_config = {
    "config_list": [
        {
            "model": "llama3",
            "base_url": "http://localhost:11434/v1",
            "api_key": "ollama"
        }
    ],
    "temperature": 0
}


def compare_models(_):
    return "GPT-4: Accurate, Closed. Claude: Safe, Closed. Mistral: Fast, Open."


assistant = AssistantAgent(
    name="assistant",
    llm_config=llm_config,
    system_message="""
You are a knowledgeable AI assistant.

Rules:
- Do NOT invent APIs, code, or sources.
- Do NOT include code blocks unless explicitly asked.
- Do NOT fabricate model developers or ownership.
- Answer in clear, structured natural language.
- If unsure, say so.
- Be factually accurate and concise.
"""
)


user_proxy = UserProxyAgent(
    name="user",
    code_execution_config={"use_docker": False}
)


user_proxy.initiate_chat(
    assistant,
    message="Can you compare GPT-4, Claude, and Mistral?",
    max_turns=1
)
```
